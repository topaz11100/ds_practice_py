```python
def quicksort(arr, start, end):
    if end-start <= 0:
        return
    pivot = arr[(start + end)//2]
    low, high = start, end
    while True:
        while arr[low] < pivot:
            low += 1
        while arr[high] > pivot:
            high -= 1
        if arr[low] == arr[high]:
            high -= 1
        if low >= high:
            break
        arr[low], arr[high] = arr[high], arr[low]
    quicksort(arr, start, low-1)
    quicksort(arr, low+1, end)
```

```python
def mergesort(arr):
    result = []
    def recursive(start, end):
        
        p = (start + end) // 2
        recursive(start, p)
        recursive(p+1, end)
        
    recursive(0, len(arr) - 1)
    return result
    
    
```
